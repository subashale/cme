```python
from shreder import Shred, show_image, show_image_cube
import cv2
import numpy as np
```

```python
path = "apple.png"
img = cv2.imread(path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
sh = Shred(img)
```

```python
# show image
show_image(img, tricks=True)
```

### default / rectangle / horizontal and vertical shred

```python
# horizontal split is default setting
appleSh = sh.default(size=100)
show_image(appleSh)
```

```python
# testing if shred has consider all the images

rows = np.concatenate(appleSh, axis=0)
show_image(rows)
```

```python
# horizontal split is default setting, in percentage model
appleSh = sh.default(size=10, percent=True) # size also works as percentage but keep percent as true
show_image(appleSh)
```

```python
# testing if shred has consider all the images

rows = np.concatenate(appleSh, axis=0)
show_image(rows)
```

```python
# horizontal split is default setting
appleSh = sh.default(angel="v", size=100)
show_image(appleSh, order="queue") # provide order : queue for better preview
```

```python
# testing if shred has consider all the images

cols = np.concatenate(appleSh, axis=1)
show_image(cols)
```

```python
# horizontal split is default setting, in percentage model
appleSh = sh.default(angel="v", size=10, percent=True) # size also works as percentage but keep percent as true
show_image(appleSh, order="queue")
```

```python
# testing if shred has consider all the images

cols = np.concatenate(appleSh, axis=1)
show_image(cols)
```

### cube mode / chunk based shred

```python
# cube with size parameter
appleSh = sh.cubic(size=(200, 100)) # h*w
show_image_cube(appleSh)
```

```python
# cube test percentage, height percentage * width percentage
appleSh = sh.cubic(size=(50, 20), percent=True) # h*w
show_image_cube(appleSh)
```

```python
# testing if shred has consider all the images

cols = [np.concatenate(col, axis=1) for col in appleSh]
show_image(np.concatenate(cols, axis=0))
```
